```python
import pandas as pd
data = pd.read_csv('Coffee_company.csv')
print(data)
data.columns = data.columns.str.strip()
data.head()
```

```python
data['Units sold']= data['Units Sold'].astype(int)
data.head()
```

```python
print(data.columns)
data.rename(columns={'Month Name':'Month'}, inplace = True)
data.head()
```

```python
data['Sales'] = data['Sales'].astype(str).str.replace(',', '', regex=False).str.replace('$', '', regex=False)

# Convert to numeric (handling errors by coercing invalid values to NaN)
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')

# Print first 5 rows for verification
print(data.head())

# Calculate and print total sales
total_sales = data['Sales'].sum()
print("Total Sales: $", total_sales)
```
